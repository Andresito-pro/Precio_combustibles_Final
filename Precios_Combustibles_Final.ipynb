```python
column_rename_map = {
    "departamentocodigo": "codigo_departamento",
    "departamentonombre": "nombre_departamento",
    "municipiocodigo": "cogigo_municipio",
    "municipionombre":	"nombre_municipio",
    "bandera": "estacion"
}
precio_com_renombrado = precio_com.rename(columns=column_rename_map)
```

##Preguntas:
Cual es la variación de precios en las diferentes estaciones de gasolina?

Hay variaciones de ventas de los diferentes productos?

```python
precio_estacion= precio_com_renombrado.groupby('producto')['precio'].mean()

precio_estacion = precio_estacion.sort_values(ascending=False).head(6)
plt.figure(figsize=(12, 6))
precio_estacion.plot(kind='bar', color='#FFCC33')
plt.title("Top 5 tipos de Combustibles")
plt.xlabel("producto")
plt.ylabel("precio")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##Analisis gráfica 1:
Se observan los diferentes tipos de combustibles que se han trabajado y ahora están graficados.

Las variaciones en los costos se pueden deber a factores como producción, demanda y disponibilidad.

```python
precio_com_renombrado.groupby(["estacion"]) [["producto"]].max()
```

```python
pd.crosstab(precio_com_renombrado.estacion, precio_com_renombrado.producto)
```

##Conclusiones de la tabla 1
En  la columna 1 se encuentran los nombres de las estaciones de gasolina y en las siguientes filas se muestra la cantidad de sedes que comercilizan los diferentes tipos de combustible.

```python
precio_com_renombrado['precio'] = pd.to_numeric(precio_com_renombrado['precio'], errors='coerce')
precio_tipos = precio_com_renombrado.groupby('estacion')['precio'].sum()
precio_tipos= precio_tipos.sort_values(ascending=False)
plt.figure(figsize=(18, 10))
precio_tipos.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Diferencia de precios")
plt.xlabel("estacion")
plt.ylabel("precio")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##Conclusion gráfica 2:
Las barras más altas representan estaciones con mayores ventas totales de combustible.
 Esto puede indicar que estas estaciones tienen una mayor demanda o que venden combustibles a precios más altos.


```python
precio_acpm_diesel = precio_com_renombrado[precio_com_renombrado['producto'] == 'ACPM - DIESEL']
tabla_precios = precio_acpm_diesel.groupby('estacion')['precio'].mean().reset_index()
tabla_precios
```

```python
precio_acpm = precio_com_renombrado[precio_com_renombrado['producto'] == 'ACPM - DIESEL']
tabla_precios_acpm = precio_acpm.groupby('estacion')['precio'].mean().reset_index()

print(tabla_precios_acpm)
plt.figure(figsize=(12, 6))
plt.bar(tabla_precios_acpm['estacion'], tabla_precios_acpm['precio'], color='yellow')
plt.title('Precio promedio de Gasolina ACPM por estación')
plt.xlabel('Estación')
plt.ylabel('Precio promedio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
